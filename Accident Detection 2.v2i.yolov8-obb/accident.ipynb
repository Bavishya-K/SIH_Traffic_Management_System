```python
from ultralytics import YOLO
import cv2
import os

# Load your trained model (replace with the path to your `best.pt`)
model = YOLO("D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/runs/detect/accident_detection_extended/weights/best.pt")

# Path to the folder containing test images
test_image_folder = "D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/test/images"  # Update with your test images folder path
output_folder = "D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/output"  # Folder to save annotated images

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Process each test image
for image_name in os.listdir(test_image_folder): 
    image_path = os.path.join(test_image_folder, image_name)

    # Load the image
    image = cv2.imread(image_path)

    # Run inference on the image
    results = model(image, save=False)  # `save=False` avoids saving directly via YOLO

    # Extract prediction information
    boxes = results[0].boxes  # Bounding boxes
    classes = results[0].names  # Class names from training
    detected_classes = [classes[int(cls)] for cls in boxes.cls]

    # Check if "accident" or "no accident" is detected
    if "accident" in detected_classes:
        print(f"Image '{image_name}' -> Prediction: Accident")
    elif "noaccident" in detected_classes:
        print(f"Image '{image_name}' -> Prediction: No Accident")
    else:
        print(f"Image '{image_name}' -> Prediction: Unknown")

    # Annotate the image with predictions
    annotated_image = results[0].plot()

    # Save the annotated image to the output folder
    output_path = os.path.join(output_folder, image_name)
    cv2.imwrite(output_path, annotated_image)

    # Optionally display the image for 1 second and move to the next one
    cv2.imshow("Prediction", annotated_image)
    cv2.waitKey(2000)  # Wait for 2 second before moving to the next image

# Close all OpenCV windows after processing all images
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import os

# Load your trained model (replace with the path to your `best.pt`)
model = YOLO("D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/runs/detect/accident_detection_extended/weights/best.pt")

# Path to the input video
input_video_path = "D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/test_video.mp4"  # Replace with your video path
output_video_path = "D:/Traffic_Accident/annotated_video.mp4"  # Path to save the annotated video

# Load the video
cap = cv2.VideoCapture(input_video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))  # Frames per second
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # Frame width
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Frame height
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output video

# Create a VideoWriter to save the output video
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit if no frames left

    # Run inference on the frame
    results = model(frame, save=False)  # `save=False` avoids direct saving

    # Annotate the frame with predictions
    annotated_frame = results[0].plot()

    # Display the frame (optional)
    cv2.imshow("Video Prediction", annotated_frame)

    # Write the annotated frame to the output video
    out.write(annotated_frame)

    # Break the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import os

# Load your trained model (replace with the path to your best.pt file)
model = YOLO("D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/runs/detect/accident_detection_extended/weights/best.pt")

# Folder containing test videos
test_videos_folder = "D:/Traffic_Accident/Accident Detection 2.v2i.yolov8-obb/test_videos"  # Replace with your test videos folder path
output_videos_folder = "D:/Traffic_Accident/annotated_videos_folder"  # Folder to save annotated videos

# Create the output folder if it doesn't exist
if not os.path.exists(output_videos_folder):
    os.makedirs(output_videos_folder)

# Loop through all video files in the test folder
for video_name in os.listdir(test_videos_folder):
    if video_name.endswith((".mp4", ".avi", ".mkv")):  # Ensure it's a video file
        video_path = os.path.join(test_videos_folder, video_name)
        output_video_path = os.path.join(output_videos_folder, f"annotated_{video_name}")

        # Open the input video
        cap = cv2.VideoCapture(video_path)

        # Get video properties
        fps = int(cap.get(cv2.CAP_PROP_FPS))  # Frames per second
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # Frame width
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Frame height
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output video

        # Create VideoWriter to save annotated video
        out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

        # Process the video frame by frame
        while cap.isOpened():
            ret, frame = cap.read()  # Read a frame
            if not ret:
                break  # Stop if no frames are left

            # Perform inference on the current frame
            results = model(frame, save=False)  # `save=False` avoids saving via YOLO

            # Annotate the frame with predictions
            annotated_frame = results[0].plot()

            # Write the annotated frame to the output video
            out.write(annotated_frame)

            # Display the annotated frame
            cv2.imshow("Prediction", annotated_frame)

            # Add a delay (in milliseconds) for displaying the frame
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break  # Stop if 'q' key is pressed

        # Release video resources for the current video
        cap.release()
        out.release()

        print(f"Annotated video saved at: {output_video_path}")

# Destroy all OpenCV windows
cv2.destroyAllWindows()
print("All videos processed successfully!")
```
